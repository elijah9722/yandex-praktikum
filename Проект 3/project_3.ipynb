- в столбце с высотой потолков (медианным значением);
- в столбце `balcony` пропуски заменить на 0, т.к. если при заполнении собственник не указал кол-во балконов, то, скорее всего, их нет;
- в столбцах с расстоянием до центра города и до ближайшего аэропорта. Можно заполнить значениями, вручную собранными из карт;

- в столбце `floors_total` пропусков очень мало в сравнении с общим кол-вом строк, поэтому, возможно, строки с этими пропусками можно будет удалить.

Пока не ясно, как быть с пропусками в `days_exposition`. Пропуски могут означать еще активное объявление, либо это закрытое объявление, но по каким-то причинам не отмеченное как таковое.
Пропусков в `is_apartment` очень много (88%). Они могли появиться из-за того, что собственник не указывал эти данные.
Пропуски в остальных столбцах пока можно оставить как есть. Также имеется более 60% пропусков в столбцах `parks_nearest` и `ponds_nearest`. Данные о расстоянии собирались автоматически, и заполнить эти пропуски скорее всего не выйдет.

### Заполнение пропусков

Пропуски в столбце `balcony` заменим на нули:

```python
data['balcony'] = data['balcony'].fillna(0)
print('Количество пропусков в столбце balcony:', data['balcony'].isna().sum())
```

Пропуски в столбце `ceiling_height`:

```python
print(data['ceiling_height'].median())
```

Среднее значение высоты потолков в новостройках Санкт-Петербурга варьируется в пределах 2,6 - 3 м, поэтому считаю возможным такой вариант заполнения пропусков.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
print('Количество пропусков в столбце ceiling_height:', data['ceiling_height'].isna().sum())
```

Пропуски в столбцах `airports_nearest` и `cityCenters_nearest`. Для начала посмотрим, какие населенные пункты есть в данных:

```python
print('Уникальных названий в списке:', len(data['locality_name'].value_counts()))
print(data['locality_name'].unique())
```

Населенных пунктов с пропусками очень много, вручную заполнить расстояния для всех - это очень долго. Скорее всего, расстояние до центра и до аэропорта не заполенено у той недвижимости, которая находится в Лен. области. Очевидно, что цена за квартиру в области будет ниже, чем в черте города.

Строки с пропусками в `floors_total` удалим:

```python
data = data.dropna(subset=['floors_total'])
print('Количество пропусков в столбце floors_total:', data['floors_total'].isna().sum())
```

Пропуски в `kitchen_area` можно заполнить следующим образом: на основе имеющихся данных рассчиать долю площади кухни от общей площади для каждой квартиры, собирая полученные значения в отдельный столбец. Затем нужно заполнить пропуски в полученном новом столбце медианным значенем долей площади кухонь, а по ним уже заполнить пропуски в исходном столбце.

```python
data['perc_kitchen'] = data['kitchen_area'] / data['total_area']
data['perc_kitchen'] = data['perc_kitchen'].fillna(data['perc_kitchen'].median())
data['kitchen_area'] = data['perc_kitchen'] * data['total_area']
print(data['kitchen_area'].head(10))
print('Количество пропусков в столбце kitchen_area:', data['kitchen_area'].isna().sum())
```

Пропуски в столбце `living_area` заполним аналогичным образом:

```python
data['perc_living'] = data['living_area'] / data['total_area']
data['perc_living'] = data['perc_living'].fillna(data['perc_living'].median())
data['living_area'] = data['perc_living'] * data['total_area']
print(data['living_area'].head(10))
print('Количество пропусков в столбце living_area:', data['living_area'].isna().sum())
```

```python
data['locality_name'] = data['locality_name'].str.lower() # Приведем названия населенных пунктов к нижнему регистру
data['locality_name'] = data['locality_name'].str.replace('поселок', 'посёлок') # Попробуем убрать неявные дубликаты заменой слов 
data['locality_name'] = data['locality_name'].str.replace('деревня', 'посёлок')
data['locality_name'] = data['locality_name'].str.replace('село', 'поселок')
```

```python
print('Уникальных названий в списке:', len(data['locality_name'].value_counts()))
print(data['locality_name'].unique())
```

После приведения названий к нижнему регистру и замены слов деревня/село/поселок на одно слово **посёлок** уникальных названий стало 325, было 364.

### Изменение типов данных

```python
data.info()
```

Изменим тип данных в следующих столбцах:
- `last_price` - на целочисленный формат, т.к. обычно цена за квартиру указывается без копеек;
- `floors_total`, `balcony` - на целочисленный формат, т.к. число этажей или балконов не может быть дробным;
- `first_day_exposition` - дату нужно перевести в формат datetime, т.к. сейчас это просто текст.

```python
data['last_price'] = data['last_price'].astype('int') #Переводим указанные столбцы в целочисленный формат
data['floors_total'] = data['floors_total'].astype('int')
data['balcony'] = data['balcony'].astype('int')
data.info()
```

```python
 #Переводим столбец с датой в формат datetime
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
data.head() #Проверяем, что формат заменен успешно
```

### Вывод

Пропуски в датасете погли появиться по разным причинам:
- в столбце `balcony` собственник не указывал кол-во балконов, если их нет;
- в столбцах с площадью кухни / жилой площадью пропуски могли появиться в связи с тем, что объявление нужно было выложить как можно скорее и времени на измерение площади отдельных комнат у собсвенника не было. Зато везде была заполнена основная информация - общая площадь помещения;
- в столбце `is_apartment` собственник мог просто не знать, является ли данное помещение апартаментами, поэтому пропускал этот пункт при заполнении объявления;
- пропуски в столбцах с расстояниями до центра города и до аэропорта присутствуют у той недвижимости, которая находится за пределами города, поэтому автоматического заполнения расстояния не происходило;
- пропуски в столбце `days_exposition` могли появиться из-за того, что объявление либо еще не закрыто, либо по какой-то причине не было отмечено как закрытое. 

Там, где это возможно, были заполнены пропуски значений. Также был успешно заменен тип данных в тех столбцах, где это необходимо.

## Расчёты и добавление результатов в таблицу

Соотношение жилой и общей площади, а также отношение площади кухни к общей были посчитаны и добавлены в таблицу ранее (`perc_kitchen` и `perc_living`). Теперь добавим в таблицу столбец с ценой за квадратный метр:

```python
data['m2_price'] = data['last_price'] / data['total_area']
data.head() #Проверяем, что столбец создан и верны ли в нем значения
```

Далее добавляем в таблицу год, месяц и день публикации объявления:

```python
data['year'] = data['first_day_exposition'].dt.year #Создаем столбец с годом 
data['month'] = data['first_day_exposition'].dt.month #Создаем столбец с месяцем
data['day'] = data['first_day_exposition'].dt.weekday #Создаем столбец с днем недели
data.head() #Проверяем наличие новых столбцов
```



Далее добавим этаж квартиры: первый, последний или другой:

```python
def floor(floor_number):   #Функция для категоризации по этажам
    if floor_number['floor'] == 1:
        return 'первый'
    if floor_number['floor'] == floor_number['floors_total']:
        return 'последний'
    return 'другой'

data['floor_name'] = data.apply(floor, axis=1)
data.head(15) #Проверяем, что функция работает корректно
```

### Вывод

В данном пункте были произведены необходимые расчеты и добавлены требуемые в задании столбцы.

### Изучение факторов, влияющих на стоимость

```python
def graf_last_price(df, col):  #Функция для построения графиков (общая цена квартиры)
    if col == 'floor_name':
        key = 'last_price'
    else:
        key = col

    df_filtered = df.groupby(col).median()['last_price'].reset_index().sort_values(by=key)
    graf = df_filtered.plot(x=col, y='last_price', kind='bar', figsize=(12,5), grid=True)
    return graf

def graf_m2_price(df, col): #Функция для построения графиков (цена за кв. метр)
    if col == 'floor_name':
        key = 'm2_price'
    else:
        key = col

    df_filtered = df.groupby(col).median()['m2_price'].reset_index().sort_values(by=key)
    graf = df_filtered.plot(x=col, y='m2_price', kind='bar', figsize=(12,5), grid=True)
    return graf


graf_m2_price(data, 'rooms') #Сразу строим график зависимости цены кв. метра от количества комнат
plt.title('Зависимость цены от количества комнат')
plt.xlabel('Число комнат')
plt.ylabel('Цена за кв. метр')
plt.show()
```

Если рассматривать зависимость цены за метр от кол-ва комнат, то выходит странный результат. Трёхкомнатные квартиры самые дешёвые, в то время как студии (0 комнат) выходят дороже всех остальных квартир. Проверим зависимость общей цены квартиры от кол-ва комнат:

```python
graf_last_price(data, 'rooms')
plt.title('Зависимость цены от количества комнат')
plt.xlabel('Число комнат')
plt.ylabel('Цена квартиры')
plt.show()
```

Здесь прослеживается четкая зависимость: чем больше комнат, тем выше цена квартиры. Значения выше 6 являются выбросами.

```python
graf_m2_price(data, 'floor_name')
plt.title('Зависимость цены от этажа')
plt.xlabel('Этаж')
plt.ylabel('Цена за кв. метр')
plt.show()
```

Здесь зависимость цены за метр от этажа логичная: квартиры на первом этаже - самые дешевые, на последнем - дороже, чем на первом, но дешевле, чем на любом другом. Соответственно, цены на этажах, отличных от первого и последнего, выше всего.

Далее, чтобы построить график зависимости цены от расстояния до центра, необходимо категоризировать данные. Самый центр города обозначим в пределах трех км, следующая категория - от 3 до 6 км, затем от 6 до 18000 - в переделах КАД, а все, что за этой зоной - область:

```python
def oblast(row):
    if row['cityCenters_nearest'] < 3000:
        return 'центр города'
    elif row['cityCenters_nearest'] < 6000:
        return 'город'
    elif row['cityCenters_nearest'] < 18000:
        return 'в пределах КАД'
    elif row['cityCenters_nearest'] >= 18000:
        return 'область'

data['nearest_category'] = data.apply(oblast, axis=1) #Создадим новый столбец с категорией расстояния квартиры от центра
data[['cityCenters_nearest', 'nearest_category']].head(10) #Проверим, что функция сработала верно
```

```python
graf_m2_price(data, 'nearest_category')
plt.title('Зависимость цены от расстояния до центра')
plt.xlabel('Категория расстояния')
plt.ylabel('Цена за кв. метр')
plt.show()
```

По графику видно, что квартиры ближе к центру города - дороже всего, а квартиры в области - дешевле. Проверим, что покажет зависимость от полной стоимости квартиры:

```python
graf_last_price(data, 'nearest_category')
```

Здесь зависимость видна четче: квартиры в центре города - самые дорогие, затем идут квартиры в пределах 6 км от центра, затем остальные. Самые дешевые так же - в области.
